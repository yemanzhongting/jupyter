```python
import pandas as pd
import matplotlib.pyplot as plt

datelist = pd.date_range(start='1/10/20', end='1/30/20')
```

```python
out_list=[]
in_list=[]
add_list=[]

df_out=pd.read_csv(r'hubei.csv', engine='python', sep=',', encoding="utf_8_sig")
df_in=pd.read_csv(r'hubei_in.csv', engine='python', sep=',', encoding="utf_8_sig")
```

```python
rows=df_out.shape[0]#行数
columns=df_out.shape[1]
```

```python
df_out
```

```python
t=datelist.tolist()
tt=[]
for i in t:
    tt.append(str(i)[0:10].replace('-',''))
```

```python
df_add=df_out
```

```python
for i in range(rows):
    for j in range(columns):
        if j ==0 or j==1:
            pass
        else:
            df_add.iloc[i,j]=float(df_in.iloc[i,j])-float(df_out.iloc[i,j])
```

```python
df_add.to_csv('add.csv',encoding="utf_8_sig")
```
