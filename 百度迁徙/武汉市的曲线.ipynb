```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_out=pd.read_csv(r'hubei.csv', engine='python', sep=',', encoding="utf_8_sig")
df_in=pd.read_csv(r'hubei_in.csv', engine='python', sep=',', encoding="utf_8_sig")
```

```python
df_out
```

```python
_in=df_in.iloc[11,2:].values.tolist()
_out=df_out.iloc[11,2:].values.tolist()
_add=[_in[i] - _out[i] for i in range(len(_in))]
```

```python
datelist = pd.date_range(start='1/1/20', end='1/30/20')
t=datelist.tolist()
tt=[]
for i in t:
    tt.append(str(i)[0:10].replace('-',''))
```

```python
plt.figure(figsize=(8,8),dpi=300)
ax1 = plt.subplot(211)
```

```python
plt.plot(tt, _in,  color='skyblue', label='in',ls='-.')#ls或linestyl

plt.plot(tt, _add, color='green', label='add',ls=':')
plt.plot(tt, _out,   color='red', label='out',ls='steps')

plt.xticks(list(tt)[::6],tt[::6])
plt.xlabel('date')
plt.ylabel('value')
plt.legend()
```
