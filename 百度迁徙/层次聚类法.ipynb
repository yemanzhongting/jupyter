```python
df_nfr=pd.read_csv(r'resultnfr.csv', engine='python', sep=',', encoding="utf_8_sig")
```

```python
#df_noname=df_nfr.drop(['area'],axis=1)
df_noname=df_nfr.set_index('area')
df_noname.index.values.tolist()
```

```python
from scipy.cluster import hierarchy  #用于进行层次聚类，话层次聚类图的工具包
from scipy import cluster   
import matplotlib.pyplot as plt
from sklearn import decomposition as skldec #用于主成分分析降维的包
from matplotlib.font_manager import FontProperties #字体管理器
 
#设置汉字格式
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=15)

import pylab as mpl     #import matplotlib as mpl
 
#设置汉字格式
# sans-serif就是无衬线字体，是一种通用字体族。
# 常见的无衬线字体有 Trebuchet MS, Tahoma, Verdana, Arial, Helvetica,SimHei 中文的幼圆、隶书等等
mpl.rcParams['font.sans-serif'] = ['FangSong']  # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
```

```python
#开始画层次聚类树状图   
plt.figure(figsize=(12,8),dpi=300) #新建一张图进行绘制  figsize=(12,8),dpi=300 
Z = hierarchy.linkage(df_noname, method ='ward',metric='euclidean')
hierarchy.dendrogram(Z,labels = df_noname.index)

#在某个高度进行剪切
label = cluster.hierarchy.cut_tree(Z,height=0.8)
label = label.reshape(label.size,)
```

```python
#根据两个最大的主成分进行绘图
pca = skldec.PCA(n_components = 0.95)    #选择方差95%的占比
pca.fit(df_noname)   #主城分析时每一行是一个输入数据
result = pca.transform(df_noname)  #计算结果
plt.figure(figsize=(12,8),dpi=300) #新建一张图进行绘制  figsize=(12,8),dpi=300
plt.scatter(result[:, 0], result[:, 1], c=label, edgecolor='k') #绘制两个主成分组成坐标的散点图
for i in range(result[:,0].size):
    plt.text(result[i,0],result[i,1],df_noname.index[i])     #在每个点边上绘制数据名称
x_label = 'PC1(%s%%)' % round((pca.explained_variance_ratio_[0]*100.0),2)   #x轴标签字符串
y_label = 'PC2(%s%%)' % round((pca.explained_variance_ratio_[1]*100.0),2)   #y轴标签字符串
plt.xlabel(x_label)    #绘制x轴标签
plt.ylabel(y_label)    #绘制y轴标签
```

```python
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(result[:,0],result[:,1],result[:,2], s=20, c=None, depthshade=True)
for i in range(result[:,0].size):
    ax.text(result[i,0],result[i,1],result[i,2],df_noname.index[i])     #在每个点边上绘制数据名称
x_label = 'PC1(%s%%)' % round((pca.explained_variance_ratio_[0]*100.0),2)   #x轴标签字符串
y_label = 'PC2(%s%%)' % round((pca.explained_variance_ratio_[1]*100.0),2)   #y轴标签字符串
z_label = 'PC3(%s%%)' % round((pca.explained_variance_ratio_[1]*100.0),2)   #y轴标签字符串
plt.xlabel(x_label)    #绘制x轴标签
plt.ylabel(y_label)    #绘制y轴标签
ax.set_zlabel(z_label)
```
