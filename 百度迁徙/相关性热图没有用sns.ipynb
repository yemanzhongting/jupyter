```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_out=pd.read_csv(r'hubei.csv', engine='python', sep=',', encoding="utf_8_sig")
df_in=pd.read_csv(r'hubei_in.csv', engine='python', sep=',', encoding="utf_8_sig")
```

```python
df_in=df_in.drop(['code'],axis=1)#这里是axis为1代表列
df_in=df_in.set_index('area')
```

```python
df_out=df_out.drop(['code'],axis=1)#这里是axis为1代表列
df_out=df_out.set_index('area')
```

```python
df_in
```

```python
df_in_list=df_in.T.apply(sum).values.tolist()
df_in_list
```

```python
df_out_list=df_out.T.apply(sum).values.tolist()
df_out_list
```

```python
import numpy as np
out_=np.array(df_out_list)
in_=np.array(df_in_list)
bb=in_-out_
print(list(bb))
```

```python
add=list(bb)
areas=df_in.index.values.tolist()
data={
    'add':add,
    'area':areas
}
t_data=pd.DataFrame(data)
```

```python
t_data.set_index('area').to_csv('addcity.csv',encoding="utf_8_sig")
```
