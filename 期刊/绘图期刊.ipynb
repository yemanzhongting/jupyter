```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'result.csv', engine='python', sep=',', encoding="utf_8_sig")
```

```python
from collections import Counter
result = Counter(df['journal'].values.tolist())
```

```python
area=[]
count=[]
for k,v in result.items():
    area.append(k)
    count.append(v)
data=pd.DataFrame({'journal':area,'count':count})
```

```python
data=data.sort_values(by='count',ascending=False)
data=data.iloc[:20,:]
data
```
